```python
# Import packages
```

```python
import pandas as pd
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
# Take a dataset
```

```python
data = {
    'Name': ['John', 'Alice', 'Bob', 'Carol', 'David'],
    'Age': [30, 25, 45, 35, 50],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Salary': [60000, 55000, 70000, 65000, 80000]
}
```

```python
# Create a DataFrame from the sample dataset
```

```python
df = pd.DataFrame(data)
df
```

```python
# Selecting the features 'Age' and 'Salary' for clustering
```

```python
X = df[['Age', 'Salary']]
X
```

```python
# Perform hierarchical clustering using the Ward method and create the linkage matrix
```

```python
linkage_matrix = sch.linkage(X, method='ward')
linkage_matrix
```

```python
# Create a dendrogram to visualize the hierarchical clustering
```

```python
dendrogram = sch.dendrogram(linkage_matrix)

# Plot the dendrogram
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Employees')
plt.ylabel('Euclidean distances')
plt.show()
```
